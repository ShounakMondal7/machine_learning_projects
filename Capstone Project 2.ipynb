Problem statement : Developing a machine learning model to predict the quality of wine (categorized as "good" or "bad") based on its physicochemical properties. This involves data collection, exploratory data analysis, model training using a Random Forest Classifier, and evaluating its performance.

Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

Data Collection

```python
wine_dataset = pd.read_csv('/content/WINE_DATASET_1.csv')
```

```python
wine_dataset.head()
```

```python
wine_dataset.isnull().sum()
```

```python
wine_dataset.describe()
```

```python
sns.catplot(x='quality', data = wine_dataset, kind = 'count')
```

```python
plot = plt.figure(figsize=(5,5))
sns.barplot(x='quality', y = 'volatile acidity', data = wine_dataset)
```

```python
plot = plt.figure(figsize=(5,5))
sns.barplot(x='quality', y = 'citric acid', data = wine_dataset)
```

Correlation

1. Positive Correlation
2. Negative Correlation

```python
correlation = wine_dataset.corr()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(correlation, cbar=True, square=True, fmt = '.1f', annot = True, annot_kws={'size':8}, cmap = 'Blues')
```

```python
X = wine_dataset.drop('quality',axis=1)
```

```python
print(X)
```

```python
Y = wine_dataset['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

Model Training:

Random Forest Classifier

```python
model = RandomForestClassifier()
model.fit(X_train, Y_train)
```

Model Evaluation:
Accuracy Score

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy : ', test_data_accuracy)
```

Building a Predictive System

```python
input_data = (7.5,0.5,0.36,6.1,0.071,17.0,102.0,0.9978,3.35,0.8,10.5)
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
prediction = model.predict(input_data_reshaped)
print(prediction)
if (prediction[0]==1):
  print('Good Quality Wine')
else:
  print('Bad Quality Wine')
```
